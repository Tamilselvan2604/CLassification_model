```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd

processed_data=pd.read_csv(r"C:\Users\Tamilselvan M\Downloads\ML\DATA\natural_disasters_2024.csv")
```

```python
processed_data['Date'] = pd.to_datetime(processed_data['Date'])
```

```python
processed_data
```

```python
X = processed_data[['Location','Magnitude','Fatalities']]
y = processed_data['Disaster_Type']
```

```python
X_encoded = pd.get_dummies(X, columns=['Location'])
```

SPLIT X_TRAIN,Y_TRAIN,X_TEST,Y_TEST

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

SCALER

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)  # You can adjust the 'k' value
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print accuracy
print(f'Accuracy: {accuracy * 100:.2f}%')
```
